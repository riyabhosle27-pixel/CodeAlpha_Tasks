```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10,5)
df = pd.read_csv("netflix_titles.csv")
```

## **1.Movies v/s TV shows/series**

```python
sns.countplot(x='type', data=df, palette='Set2')
plt.title("Distribution of Movies v/s Series-TV Shows", fontsize=20)
plt.xlabel("")
plt.ylabel("Count")
plt.show()
```

# **2.Growth In Content So Far**

```python
year_data = df['release_year'].value_counts().sort_index()
plt.plot(year_data.index, year_data.values)
plt.title("Netflix-Content Growth Over Years", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

# **3. Top Countries**

```python
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='coolwarm')
plt.title("Top 10 Content-Producing Countries")
plt.xlabel("Number of Titles")
plt.ylabel("")
plt.show()
```

# **4.Genre Analysis**

```python
genres = df['listed_in'].str.split(', ', expand=True).stack()
top_genres = genres.value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette='magma')
plt.title("Top 10 Genres shown on Netflix")
plt.show()
```

# **5.Ratings**

```python
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index, palette='Set1')
plt.title("Rating Distribution")
plt.show()
```

# **6.Year v/s Type (of movies/shows)**

```python
sns.countplot(x='release_year', hue='type', data=df, palette='Set2')
plt.title("Movies vs TV Shows Over Years")
plt.xticks(rotation=90)
plt.show()
```

# **7.Watch Duration Analysis**

```python
movies = df[df['type'] == 'Movie'].copy()
movies['duration'] = movies['duration'].str.replace(' min', '').astype(float)
sns.histplot(movies['duration'], bins=30, kde=True)
plt.title("Distribution of Movie Duration")
plt.xlabel("Minutes")
plt.show()
```
